# Time Shifting

```python
df = pd.read_csv('/Users/kaden.cho/git/kadensungbincho/udemy_local/Python-for-Finance-Repo-master/05-Pandas-with-Time-Series/time_data/walmart_stock.csv',
                index_col='Date', parse_dates=True)
```

```python
df.shift(periods=1).head()
```

```python
df.shift(periods=1).tail()
```

```python
df.tshift(freq='M').head()
```

```python
df['Open'].plot(figsize=(16, 6))
```

```python
df['Open'].plot()
df.rolling(window=7).mean()['Close'].plot(figsize=(16, 6))
df.rolling(window=90).mean()['Close'].plot(figsize=(16, 6))
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA', 'Close']].plot(figsize=(16, 6))
```

```python
df['Close'].expanding(min_periods=1).mean().plot(figsize=(16, 6))
```

## Bollingard Band

```python
# Close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

# Upper = 20MA + 2 * std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

# Lower = 20MA - 2 * std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

# Close
```

```python
df[['Close: 20 Day Mean', 'Upper', 'Lower', 'Close']].tail(200).plot(figsize=(16, 6))
```
