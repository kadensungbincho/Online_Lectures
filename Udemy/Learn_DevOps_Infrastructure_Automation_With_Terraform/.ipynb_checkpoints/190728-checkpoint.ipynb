# what is terraform

- Infrastructure as code
- Automation of your infrastructure
- Keep your infrastructure in a certain state (compliant)
    - e.g. 2 web instances with 2 volumes, and 1 load balancer
- make your infrastructure auditable
    - you can keep your infrastructure change history in a version control system like GIT
- Ansible, Chef, Puppet, Slatstack have a focus on automating the installation and configuration of software 
    - keeping the machine in compliance, in a certain state
- Terraform can automate provisioning of the infrastructue itself
    - e.g. Using the AWS, DigitalOcean, Azure API
    - works well with automation software like ansible to install software after the infrastructue is provisioned
    - So, a typical example is, spin up some hardware using AWS using Terraform, and then when that hardware becomes available you can use tools like, Ansible, Chef, Puppet to then installed and configure the software on those machines
