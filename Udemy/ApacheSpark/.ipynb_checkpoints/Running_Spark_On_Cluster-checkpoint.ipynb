# Packaging And Deploying Your Application

- Make sure there are no paths to your local filesystem used in your script! That's what HDFS, S3, etc. are for.
- Pachage up your Scala project into a HAR file (using Export in the IDE)
- You can now use spark-submit to execute driver script outside of the IDE

```bash
spark-submit --class <class object that contains your main function> --jars <paths to any dependencies> --files <files you want to placed alongside your application> <your JAR file>
``` 

# Packagin with SBT

- Like Maven for Scala
- Manages your library dependency tree for you
- Can package up all of your dependencies int o a self-contained JAR
- If you have many dependencies (or depend on a library that in turn has lots of dependencies), it makes life a lot easier than passing a ton og --jars options.
- Get it from scala-sbt.org

- Your Scala source files go in the source folder
- In your project folder, create an assembly .sbt file that contains one line: 
  addSbtPlugin("com.eed3si9n" % "sbt-assembly" % "0.14.3")
  
- Check the latest sbt documentation as this will change over time. This works with sbt 0.13.11

- At the root (alongside the src and project directories) create a build.sbt file

```bash

name := "PopularMovies"

version := "1.0"

organization := "com.kadensungbincho"

scalaVersion := "2.10.6"

libraryDependencies += Seq(
    "org.apache.spark" %% "spark-core" % "1.6.1" % "provided"
)

``` 

- Say for example you need to depend on Kafka, which isn;t built into Spark. You could add 

```bash

"org.apache.spark" %% "spark-streaming-kafka" % "1.6.1",

``` 
  to your library dependencies, and sbt will automatically fetch it and everything it needs and bundle it into you JAR file.
  
- Make sure you use the correct Spark version number, and note that I did NOT use "provied" on the line.
