Hadoop : The Definitiva Guide

- Replica is a phisical data storage on a data node. There are usually several replicas with the same content on different data nodes.
- Block is a meta-information storage on a name node and provides infomation about replica's locations and their states.
Both replica and block have their own states.

- Replica
  - Finalized state : the content of the replica is frozen that means meta-information for this block on name node is aligned with all the corresponding replica's states and data. Each block of data has a version number called Generation Stamp : GS which increases when replica is on error recovery or adding process.
