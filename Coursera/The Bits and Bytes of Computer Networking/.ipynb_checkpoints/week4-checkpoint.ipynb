# Introduction to Network Services

# Why do we need DNS?

- A global and highly distributed network service that resolves strings of letters into IP addresses for you

- Domain Name : The term we use for something that can be resolved by DNS

# The Many Steps of Name Resolution

- configured for a host to operate on a network in an expected way
    - IP address
    - Subnet mask
    - Gateway for a host
    - DNS server

- five primary types of DNS servers
    - caching name servers : generally provided by ISP
        - to store known domain name lookups for a certain amount of time
    - recursive name servers : generally provided by ISP
        - to store known domain name lookups for a certain amount of time
        - perform full DNS resolution requests
    - root name servers
    - TLD name servers
    - Authoritative name servers

- All domain names in the global DNS system have a TTL or time to live 
- TTL : A value, in seconds, that can be configured by the owner of a domain name for how long a name server is allowed to cache an entry before it should discard it and perform a full resolution again

# DNS and UDP

- UDP is connectionless

1. SYN packet to the local name server on port 53 
2. local name server send a SYN/ACK packet to the user
3. local name server talks to a root name server to find out who's responsible for the food.com (four way handshake)
4. repeat 3 with TLD name servers
5. repeat 3 with authoritive name servers

# Resource Record Types

- A record : it is used to point a certain domain name at a certain IPv4
- Quad A record : IPv6
- cname record : it is used to redirect traffic from one domain name to another

- enable www.food.com food.com access
    - register 2 DNS
    - using cname
    
- MX record : mail exchange, only for mail services 
- srv record : service record,
- TXT record : text, only for associating some descriptive text with a domain name for human consumption
    - increasingly used for additional data intended for other computers to process
    - often used to communicate configuration preferences about network services that you've entrusted other organizations to handle for your domain

# Anatomy of a Domain Name

- www.google.com
    - www : sub-domain
    - google : domains, used to demarcate where control moves from a TLD name server to an authoritative name server
    - com : TLD top level domain
    
    
- FQDN - Fully qualified domain name : when you combine all of these parts together

# DNS Zones

- Allow for easier control over multiple levels of a domain

- largecompany.com to (zoning)
    - la.largecompany.com
    - pa.largecompany.com
    - sh.largecompany.com
    
- Zone files 
    - Simple configuration files that declear all resource records for a particular zone
    
    
- Start of Authority (SOA)
    - Declares the zone and the name of the name server that is authoritative for it
  
- NS records
    - Indicate other name servers that might also be responsible for this zone
    
- Reverse lookip zone files
    - These let DNS resolvers ask for an IP and get the FQDN associated with it returned
    
- PTR (Pointer resource record) : Resolves an IP to a name

# Overview of DHCP

- IP address, subnet mask, gateway, name server
- Out of these four things, three are likely the same on just about every node on 
- IP address needs to be different on every single node on the network

- DHCP : Dynamic Host Configuration Protocol
    - An application layer protocol that automates the configuration process of hosts on a network
    
- Dynamic allocation
    - A range of IP addresses is set aside for client devices and one of these IPs is issued to these devices when they request one
    
- Fixed allocation
    - Requires a manually specified list of MAC address and their corresponding IPs
    
- Network time protocol (NTP) servers
    - Used to keep all computers on a network synchronized in time

# DHCP in Action

- DHCP is an application layer protocol which means it relies on the transport, network, data link and physical layers to operate

- DHCP discovery : The process by which a client configured to use DHCP attempts to get network configuration information

- process
    - DHCPDISCOVER(broadcast) : from UDP port 0.0.0.0.:68 to DHCP server 255.255.255.255:67 through broadcast
    - DHCPORFFER(broadcast) : 192.168.1.1:67 -> 255.255.255.255:68 from DHCP server to client machine
    - DHCPREQUEST : 0.0.0.0:68(client) -> 255.255.255.255:67(DHCP)
    - DHCPACk : 192.168.1.1:67(DHCP) -> 255.255.255.255:68(client)
- DHCP lease : all configuration
